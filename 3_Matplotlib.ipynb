# 시각화 절차
1. 도화지 생성(그래프를 추가할 틀 생성)
2. 그래프 생성
3. 여러가지 출력 옵션들 설정
4. 완성된 그림 출력

```python
# 가장 기본적인 시각화 라이브러리
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(4,3)) # 그림을 그릴 도화지 생성
# 옵션 : figsize=(가로크기, 세로크기) - 도화지의 크기

# 도화지에 그림 생성
# plt.plot() - 선을 그려주는 함수
# plt.plot(배열) - 배열의 각 요소가 y축값으로 사용됨
plt.plot([1,2,3,4,3,2,1])
plt.plot(np.arange(10,20))

# 완성된 그림 출력
plt.show() # 완성된 그림을 출력
```

```python
import numpy as np
# x축 값으로 활용할 배열 생성
# 0부터 12까지 0.01을 간격으로 배열 생성
x = np.arange(0,12,0.01)

# np.sin(x) - 사인함수 적용 결과
# 위에서 생성한 숫자배열에 사인함수를 적용
y = np.sin(x)
x, y
```

```python
# 도화지 생성
plt.figure(figsize=(5,3))

# 그래프 생성 - plt.plot(배열1, 배열2)
# 첫번째 배열 - x축값
# 두번째 배열 - y축값
plt.plot(x,y)

# 옵션 설정
plt.grid() # 격자생성
plt.xlabel('time') # x축 이름
plt.ylabel('amplitude') # y축 이름
plt.title('Sinewave') # 그림 이름

# 그래프 출력
plt.show()
```

```python
plt.figure(figsize=(5,3))

# sin파
# label - 범례이름, lw - 선굵기
# label을 따로 지정하지 않으면 범례가 생성이 안된다.
plt.plot(x,y, lw=4, label='Sin') 
# cos파
plt.plot(x, np.cos(x), label='Cos')

# 옵션
plt.grid()
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Sinewave')

plt.legend() # 위에서 설정한 label기반으로 범례를 생성
plt.xlim(0,4) # plt.xlim(최소값, 최대값)
plt.ylim(-0.5,1) # plt.ylim(최소값, 최대값)

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 5, 2, 6, 4, 3, 3]
```
