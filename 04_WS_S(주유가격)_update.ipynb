```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# jupyter 노트북 내부에 그림 출력
%matplotlib inline
```

# 1. 데이터 불러오기

```python
# 상반기 주유소 판매 데이터 불러오기
# 2019년 상반기 주유소 판매가격.csv,  encoding="cp949"
first = pd.read_csv('2019년 상반기 주유소 판매가격.csv', encoding="cp949")
first
```

```python
# 하반기 주유소 판매 데이터 불러오기
# 2019년 하반기 주유소 판매가격.csv,  encoding="cp949"
second =  pd.read_csv('2019년 하반기 주유소 판매가격.csv', encoding="cp949")
second
```

```python
# 상반기 판매가격 결측치 조회
first.isnull().sum()
first.info()
```

```python
# 하반기 판매가격 결측치 조회
second.isnull().sum()
second.info()
```

# 2. 데이터 결합

```python
# 상반기와 하반기 데이터를 상하로 결합
df = pd.concat([first, second])
# concat 옵션
# axis=0 - 행으로 결합(기본값)
# axis=1 - 열로 결합
df.reset_index(drop=True, inplace=True)
df
```

# 3. 데이터 가공 및 분석

```python
# 데이터 프레임 정보 조회

df.info()
```

```python
# 상표 열의 고유값 조회
# Series.unique() - 시리즈에 들어있는 고유값들만 출력
df['상표'].unique()
```

```python
# 지역 열의 고유값 조회

df['지역'].unique()
```

```python
# 지역 열의 값을 이용하여 시와 구를 분리한 후 각각 시, 구 열을 추가
# df['컬럼'].str - 컬럼 전체에 문자열 함수 사용가능
# expand=True 쪼갠 결과를 하나씩 가져와서 사용하겠다.
df['시'] = df['지역'].str.split(' ', expand=True)[0]
df['구'] = df['지역'].str.split(' ', expand=True)[1]
df.head()
```

```python
# 시 열의 서울을 서울특별시로 변환
df['시'] = df['시'].str.replace('서울','서울특별시')
df.head()
```

```python
# 기간 열을 datetime 형식으로 변경
# format="%Y%m%d" %Y - 년 %m - 월 %d - 일
df['기간'] = pd.to_datetime(df['기간'], format="%Y%m%d")
df.head()
```

```python
df.info()
```

```python
# 기간 열을 이용하여 년, 월, 일, 요일 정보를 추출하고
# 각각 년, 월, 일, 요일 열에 추가하세요
# df['컬럼'].dt - 날짜 속성, 날짜 함수 사용 가능
df['년'] = df['기간'].dt.year
df['월'] = df['기간'].dt.month
df['일'] = df['기간'].dt.day
df['요일'] = df['기간'].dt.day_name()
df.head()
```

```python
# 휘발유 가격이 비싼 5개 판매가격을 출력하세요.
df.sort_values(by='휘발유', ascending=False).head(10)
```

```python
# 휘발유 가격이 저렴한 5개 판매가격을 출력하세요. (단 휘발유가 0원인 데이터는 제외)
df[df['휘발유']!=0].sort_values(by='휘발유').head(5)
```

```python
# 상표별 평균 휘발유 가격을 출력하세요
df.groupby(by='상표').agg(['mean'])['휘발유']
```

```python
# 상표별 평균 휘발유 가격을 시각화하세요.
# bar차트
# pandas에서 사용하는 시각화
# ser or df .plot(kind='그래프종류')
# 'bar' - 수직 막대그래프, 'barh' - 수평막대그래프
# 기본값 - 선
df.groupby(by='상표')['휘발유'].agg(['mean']).sort_values(by='mean').plot(kind='bar')
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 계산하세요.
df.groupby(by=['상표','셀프여부'])['휘발유'].agg(['mean'])
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 시각화하세요.
# bar 차트
df.groupby(by=['상표','셀프여부'])['휘발유'].agg(['mean']).plot(kind='bar')
```

```python
# 셀프여부에 따라 휘발유 가격을 boxplot으로 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
plt.figure(figsize=(12,8))
sns.boxplot(x="셀프여부", y="휘발유", data=df[df['휘발유']!=0], palette="Set1")
plt.show()
```

```python
# 휘발유 가격을 브랜드 별로 셀프와 일반 주유소의 가격을 boxplot을 이용하여 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
plt.figure(figsize=(12,8))
sns.boxplot(x="상표", y="휘발유", hue="셀프여부", data=df[df['휘발유']!=0], palette="Set1")
plt.show()
```

```python
# 구별 평균 휘발유 가격을 계산하세요
df.groupby(by='구')['휘발유'].agg(['mean']).sort_values('mean', ascending=False)
```

```python
# 구별 평균 휘발유 가격을 시각화하세요.
# 높은 가격부터 출력하세요
# bar 차트
df.groupby(by='구')['휘발유'].agg(['mean']).sort_values('mean', ascending=False).plot(kind='bar')
```

```python
# 월별 평균 휘발유 가격을 계산하세요
df.groupby(by='월')['휘발유'].agg(['mean'])
```

```python
# 월별 평균 휘발유 가격을 시각화 하세요.
# bar 차트
df.groupby(by='월')['휘발유'].agg(['mean'])
```

```python
# 구와 월 열을 이용하여 평균 휘발유 가격을 피벗 테이블로 표현하세요
df.pivot_table(index='월',columns='구', values='휘발유', aggfunc='mean')
```

```python
# 구와 월 열을 이용하여 평균 휘발유 가격을 시각화 하세요 (heatmap 이용)
plt.figure(figsize=(10,5))
sns.heatmap(df.pivot_table(index='월',columns='구', values='휘발유', aggfunc='mean'))
```
