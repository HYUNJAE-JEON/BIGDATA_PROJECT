````python
# 도화지 스타일 설정
sns.set_style('darkgrid')
# white - 흰색 배경, dark - 어두운 배경,
# whitegrid - 흰색 배경에 격자 적용, darkgrid - 어두운 배경에 격자 적용
plt.figure(figsize=(5,3))
x = np.arange(0,10,0.01)
plt.plot(x,np.sin(x))
plt.plot(x,2*np.sin(x+0.5))
plt.plot(x,3*np.sin(x+1))

sns.set_palette("deep")

# sns.despine() - 위쪽, 오른쪽 테두리를 삭제
# offset=20 - 축과 그래프를 얼마나 떼어놓을것인가.
sns.despine(offset=20)
plt.show()
````

## tips 데이터셋
#### 한 레스토랑에 방문한 고객의 정보
|total_bill|tip|sex|smoker|day|time|size|
|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
|식사비용|팁|성별|흡연여부|방문요일|방문시간대|규모|

````python
# tips 데이터셋을 불러와서
# 변수 tips에 저장(DataFrame형태로 불러온다)
tips = sns.load_dataset('tips')
tips
# 한 레스토랑에 방문한 고객의 정보
````

````python
sns.set_style('whitegrid')

plt.figure(figsize=(5,3))

# boxplot - 데이터 분포를 확인하기 좋은 시각화.
sns.boxplot(y=tips['total_bill'])

plt.show()
````

````python
plt.figure(figsize=(5,3))

# sns.boxplot(y=tips['total_bill'],x=tips['day'])

# data - 전체 데이터 셋
# x - 컬럼이름
# y - 컬럼이름
# 전체 식당 데이터에서 x축은 방문 요일
# y축은 음식값으로 박스플롯을 그린다.
sns.boxplot(data=tips,x='day',y='total_bill')
plt.show()
````

````python
plt.figure(figsize=(5,3))
sns.boxplot(y=tips['total_bill'],x=tips['day'])
plt.show()
````

````python
plt.figure(figsize=(5,3))
sns.boxplot(data=tips,x='day',y='total_bill',
           hue="smoker") # hue - 범례로 사용할 컬럼
plt.show()
````

````python
plt.figure(figsize=(5,3))
# 데이터의 분포를 점으로 출력해준다.
sns.swarmplot(data=tips,x='day',y='total_bill',
           hue="smoker")
plt.show()
````

````python
plt.figure(figsize=(5,3))
sns.boxplot(data=tips,x='day',y='total_bill')
sns.swarmplot(data=tips,x='day',y='total_bill')
plt.show()
````

````python
# sns.set() - 스타일 설정
sns.set(font_scale=1.3)
# font_scale - 글자의 크기 설정

sns.set_style('whitegrid')
plt.figure(figsize=(6,4))

# 산점도 분포 - x축값, y축값을 기준으로 점을 그려준다.
sns.scatterplot(x='total_bill',y='tip',
          data=tips, hue='smoker')

plt.show()
````

````python
plt.figure()

# 산점도의 분포를 파악하는 회귀선(추세선)을 그려주는 시각화
sns.lmplot(x='total_bill',y='tip',
          data=tips, hue='smoker',
          height=5) # height - 그림의 크기.
plt.show()
````

````python
plt.figure(figsize=(5,5))
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
````

## flights 데이터셋
#### 한 비행기의 월간 승객수
|year|mon|passengers|
|:---:|:---:|:---:|
|탑승년도|탑승월|탑승객수|

````python
flights = sns.load_dataset('flights')
flights
````

````python
flights = flights.pivot_table(
                  values='passengers',
                  index='month',
                  columns='year')
flights
````

## pivot_table과 같이쓰면 좋은 그래프
### heatmap
### 2차원 테이블 데이터의 수치의 분포를 한눈에 확인시켜주는 시각화
- 각 타일의 데이터를 색상으로 표현하여 분포를 한눈에 확인할 수 있다.

````python
plt.figure(figsize=(8,6))
sns.set(font_scale=1)
# sns.heatmap(pivot_table데이터, 
#             annot=True(값을출력) or False(생략:기본설정),
#             fmt='d' - 출력할 값의 형태:'d'-10진수,'f'-실수)
sns.heatmap(flights, annot=True, fmt='d', cmap='Blues')

plt.show()
````
